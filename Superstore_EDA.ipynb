## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

```python
df = pd.read_csv("Sample_Superstore.csv", encoding='latin1')
df.head()
```

## Checking the Basic Info

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## Data Cleaning

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day
```

## Univariate Analysis (Single Column)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], kde=True)
plt.title("Sales Distribution")
```

```python
sns.histplot(df['Profit'], kde=True)
plt.title("Profit Distribution")
```

```python
sns.countplot(df['Category'])
plt.title("Category Count")
```

## Bivariate Analysis (Two Columns)

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs Profit')
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Total Sales by Category")
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Region")
```

## Multivariate Analysis

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

## Deep-Dive Retail Insights

```python
df.groupby('Category')['Sales'].sum()
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('Region')['Profit'].sum()
```

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
```

## Time Series Analysis

```python
df.groupby('Order Date')['Sales'].sum().plot(figsize=(12,5))
plt.title('Sales Over Time')
```
